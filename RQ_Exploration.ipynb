# Question1 Which are the top 10 countries of origin for the artists of MoMAï¼Ÿ

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read the file
df = pd.read_csv('Artists.csv')

# count the number of artists of each country
nationality_counts = df['Nationality'].value_counts()

# choose the top10 countries
top_nationalities = nationality_counts.head(10)

# creating DataFrame
df_nationalities = pd.DataFrame({'Nationality': top_nationalities.index, 'ArtistCount': top_nationalities.values})

# making the plot
plt.figure(figsize=(10, 5))
sns.barplot(x='Nationality', y='ArtistCount', data=df_nationalities)
plt.title('Top 10 Countries of Origin for the Artists of MoMA')
plt.xlabel('Nationality')
plt.ylabel('Number of Artists')
plt.show()
```

#### Analysis and Result
According to the graph above, the artists of MoMA are primarily from America, Germany, Britain, France, Italy, Japan, Swisslannd, Netherland, Russia, and Austrian. Most of the countries are in America and Europe, with only one country of Asia. It refelcts the cultural communication activities between different countries within MoMA.  

# Question2 What is the birth date distribution of male, female, and non-binary artists of MoMA after the 1800sï¼Ÿ

```python
# create the a subset of gender and BeginDate 
df_subset = df[['Gender', 'BeginDate']]

# Making boxplot
sns.boxplot(x='Gender', y='BeginDate', data=df_subset)

# Adjusting the range of BeginDate into 1800 to 2023
plt.ylim(1800, 2023)

plt.title('Birth Date Distribution of Male, Female, and Non-binary Artists of MoMA after 1800s')
plt.xlabel('Gender')
plt.ylabel('Birth Year')
plt.show()
```

#### Analysis and Result
The boxplot above shows an overall trend that most of the artists of MoMA were born after 1900. 

In MoMA, the birth years of female artists are mainly concentrated between 1910 and 1956, while those of male artists are primarily concentrated between 1890 and 1945. The overall distribution suggests a later trend for female artists. This may be attributed to the significant progress of The women's liberation movement from the 1960s to the 1970s, providing more opportunities and recognition for women born in the 1910s to the 1960s in the field of art.

Significantly, the non-binary artists suddenly occurred after 1975. The emergence of non-binary artists after 1975 is linked to evolving societal attitudes towards gender diversity. Cultural shifts during the late 1970s and 1980s fostered greater acceptance and visibility for non-binary identities. Legal changes and a more inclusive art landscape further facilitated the expression and recognition of non-binary artists.
